```python
#Calcula el cambio porcentual respecto al precio anterior y agrega una nueva columna con esos valores
data['pct-change']= data['Close'].pct_change()
```

```python
data.head()
```

```python
import numpy as np
```

```python
data['Log_returns']= np.log(data['Close']/data['Close'].shift())
```

```python
data.tail()
```

```python
#calculamos la desviaciÃ³n standar diaria
data['Log_returns'].std()
```

```python
volatility = data['Log_returns'].std()*252**0.5
```

```python
volatility
```

```python
str_vol=str(round(volatility, 4)*100)
```

```python
str_vol
```

```python
fig, ax = plt.subplots()
data['Log_returns'].hist(ax=ax, bins=50, alpha=0.6, color='b')
ax.set_xlabel("Log returns")
ax.set_ylabel("Freq of log return")
ax.set_title("AAPL Volatility: " + str_vol + "%")
```

```python
#Calculamos el moving average de las ultimas 10 ruedas. 
#Si el promedio movil te da mÃ¡s bajo que el precio es que estamos en una onda alcistad
data['MA10'] = data['Close'].rolling(10).mean()
```

```python
data.tail()
```

```python
data['EMA10']= data['Close'].ewm(span=10, adjust=False).mean()
```

```python
data['MA100'] = data['Close'].rolling(100).mean()
```

```python
data.tail()
```

```python
#Graficos los EMA y los MA junto con los precios de cierre. Si el EMA estÃ  por encima del MA, ooportunidad de compra
fix, ax = plt.subplots()
data[['MA10', 'EMA10','MA100']].loc['2022-01-01':].plot(ax=ax)
data['Close'].loc['2022-01-01':].plot(ax=ax, alpha=0.5)
ax.set_title('AAPl EMA, MA10 & MA100')
```

```python
data.tail()
```

```python
#Vamos a calcular la MACD y la Signal Line. La MACD es la diferencia entre EMA12 (12 dias) y EMA26(26 dias). La Signal Line es la EMA9 (9dÃ­as)
exp1=data['Close'].ewm(span=12, adjust=False).mean()
exp2=data['Close'].ewm(span=26, adjust=False).mean()
data['MACD']= exp1 - exp2
data['Signal_Line'] = data['MACD'].ewm(span=9, adjust=False).mean()
```

```python
data.tail()
```

```python
fix, ax = plt.subplots()
data[['MACD', 'Signal_Line']].loc['2022-01-01':].plot(ax=ax)
data['Close'].loc['2022-01-01':].plot(ax=ax, alpha=0.5, secondary_y=True) #Secondary_y es para setear un eje secundario para la data que estamos marcando. 
#Con el MACD y la Signal Line: Cuando la MACD estÃ¡ arriba de la Signal line, es seÃ±al de compra. Cuando la MACD estÃ¡ abajo de la Signal Line, es momento de vender.
```

```python
ticker = "QQQ"
start = dt.datetime(2019, 1, 1)
dataq = pdr.get_data_yahoo(ticker, start)
```

```python
dataq.tail()
```

```python
expA=dataq['Close'].ewm(span=12, adjust=False).mean()
expB=dataq['Close'].ewm(span=26, adjust=False).mean()
dataq['MACD']= expA - expB
dataq['Signal_Line'] = dataq['MACD'].ewm(span=9, adjust=False).mean()
```

```python
fix, ax = plt.subplots()
dataq[['MACD', 'Signal_Line']].loc['2022-01-01':].plot(ax=ax)
dataq['Close'].loc['2022-01-01':].plot(ax=ax, alpha=0.5, secondary_y=True)
ax.set_title('QQQ')
```

```python
#Stochastic Oscillator
high14= data['High'].rolling(14).max()
low14= data['Low'].rolling(14).max()
data['%K']= (data['Close']- low14)*100/(high14-low14)
data['%D']= data['%K'].rolling(3).mean()
```

```python
data.tail()
```

```python
fix, ax = plt.subplots()
data[['%K', '%D']].loc['2022-06-01':].plot(ax=ax)
ax.axhline(-800, c='r', alpha=0.3)
ax.axhline(-200, c='r', alpha=0.3)
#Cuando %K estÃ¡ debajo de %D, es seÃ±al de venta
```
